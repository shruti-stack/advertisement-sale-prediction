*building models,selecting features*

*building first model between tv and sales*

```python
x = data[['TV']]
x
```

```python
y = data[['Sales']]
y
```

> sales = m.tv+c

*importing library to perform ml*

> sklearn - Scikit-learn (Sklearn) is the most useful and robust library for machine learning in Python. It provides a selection of efficient tools for machine learning and statistical modeling including classification, regression, clustering and dimensionality reduction via a consistence interface in Python. This library, which is largely written in Python, is built upon NumPy, SciPy and Matplotlib.



> LinearRegression - Linear Regression is a machine learning algorithm based on supervised learning. It performs a regression task. Regression models a target prediction value based on independent variables. It is mostly used for finding out the relationship between variables and forecasting.

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
```

```python
linear.fit(x,y)
```

* fit will try to find the best value for m and c i.e y=mx+c

```python
linear.intercept_ #c value
```

```python
linear.coef_ #m value
```

```python
linear.score(x,y)
```

*statsmodels*

```python
import statsmodels.formula.api as smf
lm = smf.ols(formula='Sales ~TV', data=data).fit()
lm.summary()
```

```python
lm = smf.ols(formula='Sales ~TV +Radio' , data=data).fit()
lm.summary()
```

```python
lm = smf.ols(formula='Sales ~TV +Radio +Newspaper', data=data).fit()
lm.summary()
```
